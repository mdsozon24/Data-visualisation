```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
iris = sns.load_dataset('iris')
```

```python
tips
```

```python
sns.scatterplot(x = 'total_bill',y = 'tip',data = tips,hue = 'day',size = 'size')
```

```python
sns.histplot(tips['total_bill'],kde=True,bins = 70)
```

```python
sns.barplot(x='total_bill',y='tip',data = tips)
```

```python
sns.boxplot(x='day',y='total_bill',data = tips,hue = 'sex')
```

```python
sns.stripplot(x='day',y='total_bill',data = tips,hue='sex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind = 'kde',shade = True,cmap ='YlGnBu')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind = 'hex',cmap ='YlGnBu')
```

```python
sns.pairplot(titanic.select_dtypes(['number']),hue = 'pclass')
```
